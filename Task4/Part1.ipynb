```python
import png
from random import randint, choice
from IPython.display import Image
from math import log
```

```python
def random_travel(dimension : int, root_x : int, root_y : int, iterations_count) -> list[list[bool]]:
    unsticked_points = [[True for _ in range(dimension)] for _ in range(dimension)]
    unsticked_points[root_x][root_y] = False

    for iteration in range(iterations_count):
        current_x = randint(1, dimension - 2)
        current_y = randint(1, dimension - 2)
        if(not unsticked_points[current_x][current_y]):
            iteration -= 1
            continue
        should_stick = False
        while True:
            for dx in range(-1, 2):
                for dy in range(-1, 2):
                    if(not unsticked_points[current_x + dx][current_y + dy]):
                        should_stick = True

            if(should_stick):
                break
            current_x = current_x + choice([-1, 1])
            current_y = current_y + choice([-1, 1])
            current_x = min(max(1, current_x), dimension - 2)
            current_y = min(max(1, current_y), dimension - 2)

        unsticked_points[current_x][current_y] = False



    return unsticked_points
```

```python
def show_results(dimension, root_x, root_y, iterations):
    print("Fractal Dimension: ", log(iterations) / log(dimension))
    filename = f"fractals/{dimension}_{root_x}_{root_y}_{iterations}.png"
    pixels = random_travel(dimension, root_x, root_y, iterations)
    png.from_array(pixels, 'L;1').save(filename)
    display(Image(filename=filename))
    
```

```python
show_results(100, 50, 50, 500)
```

```python
show_results(150, 30, 30, 1000)
```
